```python
# Method 2: Direct pickle load (if Method 1 fails)
import pickle

try:
    with open("../artifacts/101/101.pkl", "rb") as f:
        data_sample = pickle.load(f)
    print("✅ Loaded with pickle")
except ModuleNotFoundError as e:
    print(f"❌ Missing module: {e}")
    print("Try installing the missing module or use Method 1 (joblib)")
except Exception as e:
    print(f"❌ Error: {e}")
```
